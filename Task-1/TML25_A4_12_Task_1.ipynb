```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Load the Data ---
# Use the exact paths from your results folder
path_imagenet = './results/resnet18_25_07_21_02_50/descriptions.csv'
path_places = './results/resnet18_places365_25_07_21_01_25/descriptions.csv'

df_imagenet = pd.read_csv(path_imagenet)
df_places = pd.read_csv(path_places)

print("--- ImageNet Model Results ---")
print(df_imagenet.head())
print("\n--- Places365 Model Results ---")
print(df_places.head())
```

```python
# --- Top 10 Concepts (ImageNet) ---
top_concepts_imagenet = df_imagenet['description'].value_counts().head(10)
print("Top concepts learned by the ImageNet model:")
print(top_concepts_imagenet)

plt.figure(figsize=(10, 5))
top_concepts_imagenet.plot(kind='barh').invert_yaxis() # horizontal bar chart
plt.title('Top 10 Concepts Learned by ResNet18 on ImageNet')
plt.xlabel('Number of Neurons')
plt.show()

# --- Top 10 Concepts (Places365) ---
top_concepts_places = df_places['description'].value_counts().head(10)
print("\nTop concepts learned by the Places365 model:")
print(top_concepts_places)

plt.figure(figsize=(10, 5))
top_concepts_places.plot(kind='barh').invert_yaxis()
plt.title('Top 10 Concepts Learned by ResNet18 on Places365')
plt.xlabel('Number of Neurons')
plt.show()
```

```python
# --- Count Unique Concepts ---
unique_count_imagenet = df_imagenet['description'].nunique()
unique_count_places = df_places['description'].nunique()

print(f"The ImageNet model learned {unique_count_imagenet} unique concepts.")
print(f"The Places365 model learned {unique_count_places} unique concepts.")
```

```python
# --- Compare Layer4 vs Layer2 (for the ImageNet model) ---
# See what complex objects the last layer (layer4) learns
l4_concepts = df_imagenet[df_imagenet['layer'] == 'layer4']['description'].value_counts().head(10)
print("Top concepts in Layer 4 (ImageNet):")
print(l4_concepts)

# See what simpler textures/colors the first layer we analyzed (layer2) learns
l2_concepts = df_imagenet[df_imagenet['layer'] == 'layer2']['description'].value_counts().head(10)
print("\nTop concepts in Layer 2 (ImageNet):")
print(l2_concepts)
```
